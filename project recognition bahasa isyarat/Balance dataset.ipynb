```python
import os
import random
import shutil
from pathlib import Path
```

```python
# ============================================================
# KONFIGURASI - Sesuaikan path dataset kamu di sini
# ============================================================
SOURCE_DIR = "asl_alphabet_train"        # Folder dataset asli
OUTPUT_DIR = "dataset_alfabet"   # Folder output hasil pengurangan
IMAGES_PER_CLASS = 1000           # Jumlah gambar per class
RANDOM_SEED = 42                  # Seed biar hasilnya konsisten
# ============================================================
```

```python
SUPPORTED_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}

def reduce_dataset(source_dir, output_dir, images_per_class, seed=42):
    source_path = Path(source_dir)
    output_path = Path(output_dir)

    if not source_path.exists():
        print(f"[ERROR] Folder '{source_dir}' tidak ditemukan!")
        return

    classes = sorted([d for d in source_path.iterdir() if d.is_dir()])

    if not classes:
        print(f"[ERROR] Tidak ada subfolder class di '{source_dir}'!")
        return

    print(f"Ditemukan {len(classes)} class: {[c.name for c in classes]}")
    print(f"Target: {images_per_class} gambar per class\n")

    random.seed(seed)
    summary = []

    for cls in classes:
        all_images = [
            f for f in cls.iterdir()
            if f.is_file() and f.suffix.lower() in SUPPORTED_EXTENSIONS
        ]

        total_available = len(all_images)

        if total_available == 0:
            print(f"[SKIP] Class '{cls.name}' - tidak ada gambar ditemukan.")
            continue

        if total_available < images_per_class:
            print(f"[WARN] Class '{cls.name}' hanya punya {total_available} gambar. Semua akan disalin.")
            selected = all_images
        else:
            selected = random.sample(all_images, images_per_class)

        out_cls_path = output_path / cls.name
        out_cls_path.mkdir(parents=True, exist_ok=True)

        for img in selected:
            shutil.copy2(img, out_cls_path / img.name)

        summary.append((cls.name, total_available, len(selected)))
        print(f"[OK] Class '{cls.name}': {total_available} → {len(selected)} gambar disalin.")

    print("\n" + "="*50)
    print("RINGKASAN HASIL")
    print("="*50)
    total_before = sum(s[1] for s in summary)
    total_after  = sum(s[2] for s in summary)
    print(f"{'Class':<10} {'Sebelum':>10} {'Sesudah':>10}")
    print("-"*32)
    for name, before, after in summary:
        print(f"{name:<10} {before:>10} {after:>10}")
    print("-"*32)
    print(f"{'TOTAL':<10} {total_before:>10} {total_after:>10}")
    print(f"\nDataset hasil disimpan di: '{output_dir}'")

reduce_dataset(SOURCE_DIR, OUTPUT_DIR, IMAGES_PER_CLASS, RANDOM_SEED)
```
