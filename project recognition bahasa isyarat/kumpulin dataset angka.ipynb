```python
# =====================================================
# AUTO CAPTURE DATASET ANGKA
# Kamera otomatis simpan gambar tiap 0.3 detik
#
# Tombol:
# 0 - 9  = pilih angka
# S      = mulai capture
# P      = pause
# Q      = keluar
# =====================================================

import cv2
import os
import time

# =====================================================
# SETTING
# =====================================================
BASE_FOLDER = "dataset_angka"
TOTAL_GAMBAR = 200          # jumlah gambar per angka
INTERVAL = 0.3             # tiap 0.3 detik simpan

# =====================================================
# BUAT FOLDER 0 - 9
# =====================================================
for i in range(10):
    os.makedirs(os.path.join(BASE_FOLDER, str(i)), exist_ok=True)

# =====================================================
# CAMERA
# =====================================================
cap = cv2.VideoCapture(0)

label = None
capturing = False
last_time = time.time()

while True:

    ret, frame = cap.read()

    if not ret:
        break

    frame = cv2.flip(frame, 1)

    tampil_label = "-" if label is None else label

    # =================================================
    # HITUNG FILE DI FOLDER
    # =================================================
    jumlah = 0

    if label is not None:
        folder = os.path.join(BASE_FOLDER, label)
        jumlah = len(os.listdir(folder))

    # =================================================
    # AUTO SIMPAN
    # =================================================
    if capturing and label is not None:

        now = time.time()

        if now - last_time >= INTERVAL:

            if jumlah < TOTAL_GAMBAR:

                filename = os.path.join(
                    BASE_FOLDER,
                    label,
                    f"{jumlah+1}.jpg"
                )

                cv2.imwrite(filename, frame)

                last_time = now

            else:
                capturing = False
                print(f"✅ Angka {label} selesai {TOTAL_GAMBAR} gambar")

    # =================================================
    # TAMPILAN
    # =================================================
    cv2.putText(frame, f"ANGKA : {tampil_label}",
                (10,40),
                cv2.FONT_HERSHEY_SIMPLEX,
                1, (0,255,0), 2)

    cv2.putText(frame, f"JUMLAH: {jumlah}/{TOTAL_GAMBAR}",
                (10,80),
                cv2.FONT_HERSHEY_SIMPLEX,
                1, (0,255,255), 2)

    status = "CAPTURE ON" if capturing else "PAUSE"

    cv2.putText(frame, status,
                (10,120),
                cv2.FONT_HERSHEY_SIMPLEX,
                1, (0,0,255), 2)

    cv2.putText(frame,
                "0-9 Pilih | S Start | P Pause | Q Quit",
                (10,460),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.6, (255,255,255), 2)

    cv2.imshow("Auto Capture Dataset Angka", frame)

    # =================================================
    # KEYBOARD
    # =================================================
    key = cv2.waitKey(1) & 0xFF

    # pilih angka
    if key >= ord('0') and key <= ord('9'):
        label = chr(key)
        capturing = False

    elif key == ord('s'):
        if label is not None:
            capturing = True
            last_time = time.time()

    elif key == ord('p'):
        capturing = False

    elif key == ord('q'):
        break

# =====================================================
# CLOSE
# =====================================================
cap.release()
cv2.destroyAllWindows()
```
